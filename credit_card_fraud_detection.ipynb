```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card= pd.read_csv("C:/Users/Siddhartha/Downloads/Codsoft_CreditCard-Fraud_detection-main/creditcard.csv")
```

```python
credit_card.info()
```

```python
credit_card.head()
```

```python
#Checking if null values are present
credit_card.isnull().sum()
```

```python
#Dropping null values
credit_card.dropna(how='any', inplace=True)
```

```python
credit_card.isnull().sum()
```

```python
credit_card['Class'].value_counts()
```

```python
#Handling Class Imbalance
legit=credit_card[credit_card.Class==0]
fraud=credit_card[credit_card.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card.groupby('Class').mean()
```

```python
# UNDERSAMPLING
legit_sample=legit.sample(n=982)
dataset=pd.concat([legit_sample,fraud])
dataset.head()
```

```python
dataset['Class'].value_counts()
```

```python
dataset.groupby('Class').mean()
```

```python
x=dataset.drop(columns='Class',axis=1)
y=dataset['Class']
x
```

```python
y
```

```python
#Splitting into Test and Train Dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x_train.shape,x_test.shape,x.shape)
```

```python
# MODEL TRAINING
model= LogisticRegression()
print(model)
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_pred,y_train)
```

```python
print("Accuracy of the model:",training_data_accuracy*100)
```

```python
# LET US CALUCULATE THE ACCURACY SCORE ON TESTING DATA SAMPLE
x_test_pred=model.predict(x_test)
testing_accuracy=accuracy_score(x_test_pred,y_test)
print("Accuracy of the model with testing Data:",testing_accuracy*100)
```

```python
# CALCULATING PRECISION, RECALL AND F1 SCORE
from sklearn.metrics import precision_score, recall_score,f1_score

#PRECISION

precision = precision_score(x_test_pred,y_test)
print("Precision of the model:",precision)

#RECALL

recall = recall_score(x_test_pred,y_test)
print("Recall of the model:",recall )

#F1

f1 = f1_score(x_test_pred,y_test)
print("F1 Score of the model:",f1)
```
